```python
from classifier_manager import *
from dataset_manager import *
from metrics import *
import matplotlib.pyplot as plt
from roc_pr_tools import *
from plot_tools import *

%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
n = 100
y_test = np.random.randint(2, size = n)
output = np.random.uniform(0.0, 1.0, size = n)
```

```python
fpr, tpr = get_roc_curve(output, y_test)
print("Area under ROC curve: ", get_auc(fpr, tpr))

recalls, precisions = get_pr_curve(output, y_test)
print("Area under PR curve: ", get_auc(recalls, precisions ))
```

```python
plt.rcParams['figure.figsize'] = [12, 7]

showPlots([{'data': tpr, 'x': fpr, 'label': "ROC"},
           {'data': precisions, 'x': recalls, 'label': "PR"}])
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve, auc

precision_lib, recall_lib, thresholds_lib = precision_recall_curve(y_test, output)
print("Area under PR curve: ", -get_auc(recall_lib, precision_lib))

fp_rate, tp_rate, _ = roc_curve(y_test, output)
print("Area under ROC curve: ", get_auc(fp_rate, tp_rate))
```

```python
plt.rcParams['figure.figsize'] = [12, 7]

showPlots([{'data': tp_rate, 'x': fp_rate, 'label': "ROC"},
           {'data': precision_lib, 'x': recall_lib, 'label': "PR"}])
```
